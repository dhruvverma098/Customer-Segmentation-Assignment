```python
encoder = LabelEncoder()

df['MembershipLevel'] = encoder.fit_transform(df['MembershipLevel'])

df['Category'] = encoder.fit_transform(df['Category'])

df['Country'] = encoder.fit_transform(df['Country'])

df['PaymentMethod'] = encoder.fit_transform(df['PaymentMethod'])
```

```python
X = df[['CustomerAge',
        'Quantity',
        'UnitPrice',
        'DiscountPercent',
        'LineTotal']]
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
wcss = []

for i in range(1,11):

    kmeans = KMeans(
        n_clusters=i,
        random_state=42,
        n_init=10
    )

    kmeans.fit(X_scaled)

    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,5))

plt.plot(range(1,11),
         wcss,
         marker='o')

plt.title("Elbow Method")

plt.xlabel("Number of Clusters")

plt.ylabel("WCSS")
plt.savefig("elbow_method",dpi= 300)
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

clusters = kmeans.fit_predict(X_scaled)

df['Cluster'] = clusters
```

```python
score = silhouette_score(
    X_scaled,
    clusters
)

print("Silhouette Score:", score)
```

```python
df.groupby('Cluster')[[
    'CustomerAge',
    'Quantity',
    'UnitPrice',
    'LineTotal'
]].mean()
```

```python
pca = PCA(n_components=2)

pca_data = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    pca_data[:,0],
    pca_data[:,1],
    c=clusters
)

plt.title("Customer Segments")

plt.xlabel("PCA 1")

plt.ylabel("PCA 2")
plt.savefig("customer_segments",dpi= 300)
plt.show()
```
